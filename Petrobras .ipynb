```python
df = pd.read_csv("PETR4.SA.csv", parse_dates=True, index_col='Date')
```

```python
# 0 is the initial value, 51 is the final value 
# (last value is not taken) and 5 is the difference 
# of values between two consecutive ticks 
plt.figure(figsize=(14,6))
plt.plot(df['Close'])
plt.style.use("ggplot")
plt.xlabel("Ano-mês")
plt.ylabel("Preço (R$)")
plt.title("Evolução PETR4 (2018-2020)")
plt.xticks(rotation=60)
plt.gcf()
plt.savefig("petr4_dois_anos.jpg", dpi=100)
plt.show() 
```

```python
df2020 = df.loc['2020-01-01':'2020-03-20', 'Close']
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
df2020.plot(ax=ax)

# Add labels to the plot
style = dict(size=10, color='gray')

ax.text('2020-1-1', 29, "R$ 31,00 - 02 JAN", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))

ax.text('2020-3-11', 13, "R$ 12,00 - 20 MAR", **style,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))

ax.text('2020-2-18', 27, "Feriado Carnaval", **style, 
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))

# Label the axes
ax.set(title='Variação de preço PETR4 (2020)', ylabel='Variação de preço (R$)', xlabel="2020")
plt.xlabel("")

plt.gcf()
plt.savefig("petr_2020-3.jpg", dpi = 100)
plt.show()
```
